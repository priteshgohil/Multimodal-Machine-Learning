```python
import numpy as np
import matplotlib.pyplot as plt
import os
from sklearn.preprocessing import normalize

def normalize_multichannel(np_array_input):
    return np_array_input / np.linalg.norm(np_array_input)
```

# snatched x01

```python
wrench = np.load('data/snatched_x01/wrench_resampled.npy')
wrench = wrench[150:180, 1:] #Remove time stamp i.e. first column
for i in range(wrench.shape[1]):
    plt.plot(wrench[:,i])
plt.show()
normalized_wrench = normalize_multichannel(wrench)
for i in range(normalized_wrench.shape[1]):
    plt.plot(normalized_wrench[:,i])
plt.show()
```
