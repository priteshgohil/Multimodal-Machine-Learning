1. precision - What proportion of positive identifications was actually correct?
2. recall - What proportion of actual positives was identified correctly?

data visualization : https://www.machinelearningplus.com/plots/top-50-matplotlib-visualizations-the-master-plots-python/

```python
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
def calculate_parameters(tp, tn, fp, fn):
# def calculate_parameters(tn, tp, fn, fp):
#     accuracy = 'accuracy'
#     errorRate = 'EER'
#     prec = 'precision'
#     sensitivity = 'recall'
#     fscore = 'fscore'
    accuracy = 'A'
    errorRate = 'EER'
    prec = 'P'
    sensitivity = 'R'
    fscore = 'f1'
    result = {}
    
#     #Check if user didn't make mistake in writing result numbers
#     if(tp+tn+fp+fn)!=4750 and (tp+tn+fp+fn)!=4751:
#         raise ValueError("expected {} samples, but received {}".format(4750, (tp+tn+fp+fn)))
    
    #calculate Accuracy
    t = (tp+tn)/(tp+tn+fp+fn)
    result[accuracy] = t
    
    #calculate Error Rate
    t = (fp+fn)/(tp+tn+fp+fn)
#     result[errorRate] = t
    
    #calculate Precision = true positive / total predicted positive
    precision = (tp)/(tp+fp)
    result[prec] = precision
    
    #calculate sensitivity/ recall = true positive / total actual positive
    recall = tp/(tp+fn)
    result[sensitivity] = recall
    
    #calculate f-score
    fscr = 2*(precision*recall)/(precision+recall)
    result[fscore]=fscr
    
    return result
```

# Visualize learning curve

```python
def read_csv_file(filename):
    with open(filename) as csv_file:
        readCSV = csv.reader(csv_file, delimiter=',')
        x=[]
        y=[]
        for row in readCSV:
            if(row[1].isdigit() and row[2].isdigit):
                x.append(row[1])
                y.append(row[2])
    return x,y

def read_txt_file(filename):
    y = np.loadtxt(filename,dtype="float")
    x = np.arange(len(y))
    return x,y

def plot_with_validation_err(train_err, test_err, save_imagename,ymin=0, ymax=0.7):
    xm = 5
    ym=0.05
    color = ["black", "red", "dodgerblue", "green", "saddlebrown", "orange", "blue"]
    if(len(train_err) != len(test_err)):
        raise ValueError("Length of train_err and test_err is not equal")
    plot_lines = []
#     fig = plt.figure(figsize=(7,10))
    fig = plt.figure(figsize=(5,8))
    # plt.axis('equal')
    plt.xlabel('number of epochs')
    plt.ylabel('error per epoch')
    plt.yticks(np.arange(ymin, ymax, 0.05))
    train_legend = []
    test_legend = []
    for i,item in enumerate(train_err.items()):
        legend, xy = item 
        x,y = xy
        train_legend.append(plt.plot(x, y, color=color[i], linewidth=2, label=legend)[0])
    for i,item in enumerate(test_err.items()):
        legend, xy = item 
        x,y = xy
        if(i == 0):
            test_legend.append(plt.plot(x, y, '-.', color=color[i], linewidth=2, label="Testing_error")[0])
        else:
            plt.plot(x, y, '-.', color=color[i], linewidth=2)
            
    lab1 = [ h.get_label() for h in train_legend]
    lab2 = [ h.get_label() for h in test_legend]

    #Jugad
    lab2.append('Training_error')
    test_legend.append(train_legend[0])
    
    leg1 = plt.legend(train_legend,lab1, loc=1)
    leg2 = plt.legend(test_legend,lab2, loc=2)
    plt.grid()
    plt.gca().add_artist(leg1)
    plt.gca().set_aspect(xm/ym, adjustable='box')
    fig.tight_layout()
    plt.savefig("../project-report/images/" + save_imagename)
    plt.show()
    
def plot_only_err(train_err, test_err, save_imagename,ymin=0, ymax=0.7):
    xm = 5
    ym=0.05
    color = ["black", "red", "dodgerblue", "green", "saddlebrown", "orange", "blue"]
    plot_lines = []
    fig = plt.figure(figsize=(7,10))
    # plt.axis('equal')
    plt.xlabel('number of epochs')
    plt.ylabel('error per epoch')
    plt.yticks(np.arange(0, 0.7, 0.05))
    train_legend = []
    test_legend = []
    if(len(test_err) != 0):
        if(len(train_err) != len(test_err)):
            raise ValueError("Length of train_err and test_err is not equal")
#         for i,item in enumerate(test_err.items()):
#             legend, xy = item 
#             x,y = xy
#             if(i == 0):
#                 test_legend.append(plt.plot(x, y, '-.', color=color[i], linewidth=2, label="Validation error")[0])
#             else:
#                 plt.plot(x, y, '-.', color=color[i], linewidth=2)  
#         lab2 = [ h.get_label() for h in test_legend]
#         leg2 = plt.legend(test_legend,lab2, loc=2)
    for i,item in enumerate(train_err.items()):
        legend, xy = item 
        x,y = xy
        train_legend.append(plt.plot(x, y, color=color[i], linewidth=2, label=legend)[0])

            
    lab1 = [ h.get_label() for h in train_legend]

    leg1 = plt.legend(train_legend,lab1, loc=1)
    leg2 = plt.legend([train_legend[0]],["Training_error"], loc=2)
    plt.grid()
    plt.gca().add_artist(leg1)
    plt.gca().set_aspect(xm/ym, adjustable='box')
    fig.tight_layout()
#     plt.savefig("../project-report/images/" + save_imagename)
    plt.show()
```
